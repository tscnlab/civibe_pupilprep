# Purpose of notebook

To verify assertion from developers that 0.25 s gaps and shorter will not be present in the signal. To get data, run load_and_resample with saving both raw and resampled data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import loading_utils as load
import preprocessing_utils as prep

participant_list = [200, 201, 202, 204, 205, 206, 207, 209, 210, 211, 212, 213]
```

# <a id='toc1_'></a>[Plots of first 30 s after adaptation phase for participant 200.](#toc0_)

The vertical lines are spaced every 0.25 s.

```python
data_dir = "./results/new/"
data_suffix = "_recording_data.csv"
participant_id = 200

data_path = os.path.join(data_dir, str(participant_id) + data_suffix)
data_df = pd.read_csv(
    data_path,
    usecols=[
        "Overall time Sec",
        "Stim eye - Size Mm",
        "Trial no",
        "Trial type",
        "Trial time Sec",
        "Trial phase",
        "Left - Size Mm",
        "Right - Size Mm",
        "Eye",
        "Recording id",
        "Block",
        "Test",
    ],
)
```

Gaps shorter than the plotted grid are fairly visible.

# <a id='toc2_'></a>[Histograms of time step (t_i - t_(i-1)) for raw signal without NaNs and with NaNs](#toc0_)

```python
data_df["Time diff"] = data_df["Overall time Sec"].diff() * 1000  # in ms
data_df.loc[data_df["Time diff"] < 0, "Time diff"] = (
    np.nan
)  # to avoid a negative jump from recording to recording
data_df_notna = data_df[data_df["Stim eye - Size Mm"].notna()].copy()
data_df_notna["Time diff"] = data_df_notna["Overall time Sec"].diff() * 1000
data_df_notna.loc[data_df_notna["Time diff"] < 0, "Time diff"] = np.nan

for id in data_df["Recording id"].unique():
    print(f"Time steps for participant {participant_id}, recording {id}")
    print(
        f"With nan: max: {data_df['Time diff'][(data_df['Recording id']==id)].max()},min: {data_df['Time diff'][(data_df['Recording id']==id)].min()},mean: {data_df['Time diff'][(data_df['Recording id']==id)].mean()},stdev: {data_df['Time diff'][(data_df['Recording id']==id)].std()}"
    )

    binrange_notna = [
        0,
        data_df_notna["Time diff"][(data_df_notna["Recording id"] == id)].max(),
    ]
    binrange = [0, data_df["Time diff"][(data_df["Recording id"] == id)].max()]
    f, (ax1, ax2) = plt.subplots(
        1, 2, gridspec_kw={"width_ratios": [1, 1]}, figsize=(7, 3)
    )

    print(
        f"Without nan: max: {data_df_notna['Time diff'][(data_df_notna['Recording id']==id)].max()},min: {data_df_notna['Time diff'][(data_df_notna['Recording id']==id)].min()},mean: {data_df_notna['Time diff'][(data_df_notna['Recording id']==id)].mean()},stdev: {data_df_notna['Time diff'][(data_df_notna['Recording id']==id)].std()}"
    )
    sns.histplot(
        ax=ax1,
        data=data_df_notna[(data_df_notna["Recording id"] == id)],
        x="Time diff",
        binwidth=1,
        binrange=binrange_notna,
    )

    ax1.set_title(f"Stim eye without nans")
    ax1.set_xlabel("Time step [ms]")
    sns.histplot(
        ax=ax2,
        data=data_df[data_df["Recording id"] == id],
        x="Time diff",
        binwidth=1,
        binrange=binrange,
    )
    ax2.set_title("Stim eye with nans")
    ax2.set_xlabel("Time step[ms]")
    f.suptitle(f"Recording {id},participant {participant_id}")
    plt.tight_layout()

    plt.show()
```

Look for instance at recording 10. With NaN samples included, the maximum time step is 63 ms long, with NaN samples removed, it's at 210 ms, which is 0.21 s. There are multiple cases like these, which indicate that the without-NaN time step was indeed due to a gap with multiple NaN samples, and not a sampling time issue. That means that there are gaps shorter than 0.25 s in the signal. It is also visible in the signal plots, where there are gaps visibly shorter than 0.25 s. Therefore, I'm not sure if it's indeed the case that the algorithm was applied the way it was written out, as in that smaller than 0.25 s gaps will be filled and any gaps larger than that are registered as NaN. Where can the smaller than 0.25 s gaps in the signal come from then?

The issue is even better visible in data resampled to 30 Hz which we estimated as average true sampling rate for one eye. 

# <a id='toc3_'></a>[Histograms of NaN sequence lengths for stimulated eye in data resampled to 30 Hz](#toc0_)

```python
def find_consecutive_nans(trial):
    # find and make a series of found nan sequences
    nan_list = (
        trial["Stim eye - Size Mm"]
        .isnull()
        .astype(int)
        .groupby(trial["Stim eye - Size Mm"].notnull().astype(int).cumsum())
        .sum()
    )
    return nan_list


def mark_nan_sequences(trial):
    # count the nans, preserving the indexing of the trial (e.g. for later finding out, in which phase the found long sequences areS)
    nan_list_index_preserved = (
        trial["Stim eye - Size Mm"]
        .isnull()
        .astype(int)
        .groupby(trial["Stim eye - Size Mm"].notnull().astype(int).cumsum())
        .cumsum()
    )
    return nan_list_index_preserved
```

```python
data_dir = "./results/resampled/"
data_suffix = "_nonan_30_resampled_data.csv"

resampled_marked_df_list = []
fig, axs = plt.subplots(2, 6, figsize=(20, 6))
for j, participant_id in enumerate(participant_list):
    data_path = os.path.join(data_dir, str(participant_id) + data_suffix)
    data_df = pd.read_csv(data_path)
    data_df["NaN - Count"] = pd.Series()
    nan_sequence_lengths = []
    for i, trial_no in enumerate(sorted(data_df["Trial no"].unique())):
        trial = data_df[data_df["Trial no"] == trial_no]
        nan_list = find_consecutive_nans(trial)
        nan_list = nan_list[nan_list > 0]
        data_df.loc[data_df["Trial no"] == trial_no, "NaN - Count"] = (
            mark_nan_sequences(trial)
        )
        nan_sequence_lengths.append(nan_list)

    resampled_marked_df_list.append(data_df)
    nan_lengths = pd.concat(nan_sequence_lengths)
    nan_sequence_df = pd.DataFrame()
    nan_sequence_df["NaN duration Sec"] = nan_lengths * (1 / 30)
    sns.histplot(
        nan_sequence_df, x="NaN duration Sec", binwidth=0.1, ax=axs[j // 6, j % 6]
    )
    axs[j // 6, j % 6].set_title(str(participant_id))
    axs[j // 6, j % 6].set_xlim([0, 0.5])
    if j != 0 or j != 6:
        axs[j // 6, j % 6].set_ylabel("")

fig.suptitle(
    "Histograms of NaN sequence durations of 1 or more samples for data resampled to 30 Hz after NaN removal"
)
plt.tight_layout()
plt.show()
```

For 30 Hz resampled signal, gaps in the 0.1 s bin in the stimulated eye are the most numerous. Again, this points to the fact that missing data below 0.25 s was not filled.
