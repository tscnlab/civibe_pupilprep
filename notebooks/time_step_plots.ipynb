```python
import sys

sys.path.insert(
    1, "..\\utilities\\"
)  # adds utilities folder to path so we can import modules from it, won't be needed after packaging

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import loading_utils as load
```

```python
participant_id = 200
data_dir = './results'
data_path = os.path.join(data_dir,str(participant_id)+'_full_data.csv')

data_df = pd.read_csv(data_path)
data_df['Time diff'] = data_df['Overall time Sec'].diff() *1000 #in ms
data_df.loc[data_df['Time diff']<0,'Time diff'] = np.nan #to avoid a negative jump from session to session




for id in data_df['Session id'].unique():
    print(f'Time steps for participant {participant_id}, session {id}')
    print(
            f"With nan: max: {data_df['Time diff'][(data_df['Session id']==id)].max()},min: {data_df['Time diff'][(data_df['Session id']==id)].min()},mean: {data_df['Time diff'][(data_df['Session id']==id)].mean()},stdev: {data_df['Time diff'][(data_df['Session id']==id)].std()}"
        )
    
    binrange = [0,data_df['Time diff'][(data_df['Session id']==id)].max()]
    f,(ax1,ax2)=plt.subplots(1,2,gridspec_kw={'width_ratios': [1, 1]},figsize=(7,3))
    if data_df.loc[data_df['Session id']==id,'Eye'].values[0]=='L':
        print(
            f"Without nan: max: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Left - RadiusB Px'].notna())].max()},min: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Left - RadiusB Px'].notna())].min()},mean: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Left - RadiusB Px'].notna())].mean()},stdev: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Left - RadiusB Px'].notna())].std()}"
        )
        sns.histplot(ax=ax1,data = data_df[(data_df['Session id']==id) &(data_df['Left - RadiusB Px'].notna())], x = 'Time diff',binwidth=1,binrange=binrange)
  
    else:
        print(
            f"Without nan: max: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Right - RadiusB Px'].notna())].max()},min: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Right - RadiusB Px'].notna())].min()},mean: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Right - RadiusB Px'].notna())].mean()},stdev: {data_df['Time diff'][(data_df['Session id']==id) &(data_df['Right - RadiusB Px'].notna())].std()}"
        )
        sns.histplot(ax=ax1,data = data_df[(data_df['Session id']==id) &(data_df['Right - RadiusB Px'].notna())], x = 'Time diff',binwidth=1,binrange=binrange)
        
    
    ax1.set_title(f'Stim eye without nans')
    ax1.set_xlabel('Time step [ms]')
    sns.histplot(ax=ax2,data = data_df[data_df['Session id']==id], x = 'Time diff',binwidth=1,binrange=binrange)
    ax2.set_title('Stim eye with nans')
    ax2.set_xlabel('Time step[ms]')
    f.suptitle(f'Session {id},participant {participant_id}')
    plt.tight_layout()
    
    plt.show()
    
    
```
